```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = []
y = []
with open("test_results/size/dedup_avg_time.txt", "r") as f:
    file_data = f.read()
    file_data = file_data.split('\n')
    for couple in file_data:
        couple = couple.split()
        x.append(int(couple[0]))
        y.append(float(couple[1]))

plt.figure(figsize=(10,7))
plt.plot(x, y, 'g', linewidth=2.0)
plt.plot(x, y, 'g.', markeredgewidth=3.0)
plt.xlabel('Размер сегмента')
plt.ylabel('Время дедупликации (секунды)')
plt.show()
```
